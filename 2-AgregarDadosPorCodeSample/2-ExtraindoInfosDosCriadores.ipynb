```python
from github import Github
g = Github("Chave")
import pandas as pd
```

```python
samples = pd.read_csv("1-ExtrairDadosDePR/aws.csv")
```

```python
#del samples["New Column Title"]
del samples["Unnamed: 0"]
```

```python
samples["framework"].size
```

```python
nome_do_repositorio = ""
repositorio = object
for line in range(18500,samples["framework"].size):
  print(g.get_rate_limit())
  print(f"{line} / {samples['framework'].size}")
  pull = samples.iloc[line]
  link_quebrado = pull["link"].split("/")
  numero_do_pull = link_quebrado.pop(len(link_quebrado)-1)
  sample = link_quebrado.pop(len(link_quebrado)-2)
  owner = link_quebrado.pop(len(link_quebrado)-2)
  nome_repositorio =  owner + "/" + sample
  if (nome_do_repositorio != nome_repositorio):
    nome_do_repositorio = nome_repositorio
    #repositorio = g.get_repo(nome_do_repositorio)
  
  pull_objeto = g.get_repo(nome_do_repositorio).get_pull(int(numero_do_pull)) #repositorio.get_pull(int(numero_do_pull))
  user = pull_objeto.user
  samples.loc[line, "creator user cargo"] = user.role
  samples.loc[line, "creator user empresa"] = user.company
  samples.loc[line, "creator user local"] = user.location
  samples.loc[line, "creator user login"] = user.login
  samples.loc[line, "creator user email"] = user.email
  samples.loc[line, "creator user seguidores"] = user.followers
  samples.loc[line, "creator user tempo no GitHub"] = (pull_objeto.created_at - user.created_at).days
```

```python
samples.to_csv("1-ExtrairDadosDePR/awsAtualizado.csv", index=False)
```
